y - response variable

f(x) = B0+B1*x

yi = f(xi) + Ei

Ei = yi - f(xi)

Ei = yi -(B0+B1*xi)

B0 = mean(y) - B1*mean(x)

Sxy -> sum of cross deviations of x and y

Sxx -> sum of squared deviation of x 

Predicted value of y yi' = B0+B1*(xi)

plot x and y using scatter plot and x and y' in a line plot

```python
import matplotlib.pyplot as plt
x = [i for i in range(10)]
y = [1,3,2,5,7,8,8,9,10,12]
plt.scatter(x,y)
```

```python
mean_x = sum(x)/len(x)
mean_y = sum(y)/len(y)

print(mean_x, mean_y)
```

```python
import numpy as np
x = np.array(x)
y = np.array(y)
```

```python
print(x,y)
```

```python
B1 = np.sum(((x-mean_x)*(y-mean_y)))/np.sum(((x-mean_x)**2))
B1
```

```python
B0 = np.sum(mean_y - (B1*mean_x))
B0
```

```python
y_predicted = B0 + B1*x
y_predicted
```

```python
Error = y-y_predicted
print(Error)
```

```python

y
```

```python
y = y_predicted+Error
plt.plot(x,y_predicted,color='blue')
plt.scatter(x,y,color="red")
```

```python
x = [i for i in range(10)]

y = [1,3,2,5,7,8,8,9,10,12]

Sxx=0
Sxy=0
print(mean_x,mean_y)
for i in range(len(x)):
    Sxy+=(x[i]-mean_x)*(y[i]-mean_y)
    Sxx+=(x[i]-mean_x)**2
print(Sxx,Sxy)
B1 = Sxy/Sxx
B0 = mean_y - B1*mean_x
print(B0,B1)
y_predicted = [B1*i + B0 for i in range(10)]
print(y_predicted)
```
